```
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/plotly/datasets/master/supermarket-sales.csv"
df = pd.read_csv(url)

print("Shape:", df.shape)
display(df.head())
display(df.isnull().sum())

# KPIs
print("Total revenue:", df['Total'].sum())
print("Average invoice:", df['Total'].mean())
print("Unique invoices:", df['Invoice ID'].nunique())

rev_city = df.groupby('City')['Total'].sum().sort_values(ascending=False)
plt.figure(); rev_city.plot(kind='bar'); plt.title("Total Revenue by City"); plt.tight_layout(); plt.show()

rev_prod = df.groupby('Product line')['Total'].sum().sort_values(ascending=False)
plt.figure(); rev_prod.plot(kind='bar'); plt.title("Total Revenue by Product Line"); plt.tight_layout(); plt.show()

if 'gross income' in df.columns:
    plt.figure(); plt.hist(df['gross income'], bins=30, edgecolor='k'); plt.title("Gross Income Distribution"); plt.tight_layout(); plt.show()
```
